```python
'''
Crie um script python simples para verificar a temperatura
de um sistema. Se for maior que 30, exiba
um alerta de superaquecimento
'''
def verificar_temperatura(temp):
    if temp > 30:
        print('Alerta: Superaquecimento!')
    else:
        print('Temperatura normal.')

temperatura_atual = float(input('Digite a temperatura atual do sistema: '))
verificar_temperatura(temperatura_atual)
```

```python
'''
Missão 1
criar um programa que organize arquivos bagunçados em pastas automaticamente,
baseando-se apenas na extesão de arquivo(ex.: .pdf, .jpg).
Utilize a biblioteca os ou shutil

'''

import os
import shutil

def organizar_arquivos(caminho_da_pasta):
    # Garante que o caminho existe
    if not os.path.exists(caminho_da_pasta):
        print(f"A pasta {caminho_da_pasta} não foi encontrada.")
        return

    # Lista todos os arquivos na pasta
    for arquivo in os.listdir(caminho_da_pasta):
        caminho_completo = os.path.join(caminho_da_pasta, arquivo)

        # Verifica se é um arquivo (e não uma pasta)
        if os.path.isfile(caminho_completo):
            # Extrai a extensão do arquivo (ex: .pdf)
            nome, extensao = os.path.splitext(arquivo)
            extensao = extensao.lower().strip('.')

            # Se o arquivo não tiver extensão, ignora ou define uma pasta padrão
            if not extensao:
                extensao = 'outros'

            # Define o caminho da nova pasta baseada na extensão
            pasta_destino = os.path.join(caminho_da_pasta, extensao)

            # Cria a pasta se ela ainda não existir
            if not os.path.exists(pasta_destino):
                os.makedirs(pasta_destino)

            # Move o arquivo para a pasta correspondente
            shutil.move(caminho_completo, os.path.join(pasta_destino, arquivo))
            print(f"Movido: {arquivo} -> {extensao}/")

# --- Teste o programa aqui ---
# Para testar no Google Colab, você pode criar uma pasta de teste ou usar a raiz:
# caminho = '/content/minha_pasta_baguncada'
# organizar_arquivos(caminho)
```

```python
import os

# 1. Criar uma pasta de teste e arquivos fictícios
pasta_teste = '/content/pasta_baguncada'
os.makedirs(pasta_teste, exist_ok=True)

arquivos_exemplo = ['documento.pdf', 'foto.jpg', 'notas.txt', 'relatorio.pdf', 'planilha.csv', 'sem_extensao']

for nome_arq in arquivos_exemplo:
    caminho_arq = os.path.join(pasta_teste, nome_arq)
    with open(caminho_arq, 'w') as f:
        f.write('Conteúdo de teste')

print(f"Arquivos criados em: {pasta_teste}")
print("---\nIniciando organização...\n")

# 2. Executar a função de organização
organizar_arquivos(pasta_teste)

print("\n--- Organização concluída! ---")
print("Conteúdo final da pasta:")
for item in os.listdir(pasta_teste):
    caminho_item = os.path.join(pasta_teste, item)
    if os.path.isdir(caminho_item):
        print(f"Pasta: {item}/ -> contém: {os.listdir(caminho_item)}")
```
