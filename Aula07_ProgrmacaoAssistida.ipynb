#Números Primos

```python
import math

def eh_primo(n: int) -> bool:
  if n < 2:
    return False
  if n == 2:
    return True
  if n % 2 == 0: # Otimização: se for par e maior que 2, não é primo
    return False

  # Verifica divisores ímpares de 3 até a raiz quadrada de n
  # Não é necessário verificar números pares como divisores, pois já eliminamos n par.
  for i in range(3, int(math.sqrt(n)) + 1, 2):
    if n % i == 0:
      return False
  return True

n = int(input("Digite um número: "))
print(eh_primo(n))
```

##Calculadora simples

```python
# Calculadora Simples

def operacao(num1: float, num2: float, operador: str) -> float | str:
  operadores_validos = {
      '+': lambda a, b: a + b,
      '-': lambda a, b: a - b,
      '*': lambda a, b: a * b,
      '/': lambda a, b: a / b,
  }

  if operador not in operadores_validos:
    return "Erro: operador inválido! Escolha entre +, -, * ou /."

  if operador == '/' and num2 == 0:
    return "Erro: Não é possível dividir por zero!"

  return operadores_validos[operador](num1, num2)


if __name__ == "__main__":
  while True:
    try:
      n1_str = input("Digite o primeiro número (ou 'sair' para encerrar): ")
      if n1_str.lower() == 'sair':
        print("Calculadora encerrada.")
        break
      n1 = float(n1_str)
      break
    except ValueError:
      print("Entrada inválida. Por favor, digite um número válido.")

  if n1_str.lower() != 'sair':
    operador_aritmetico = input("Digite o operador aritimético desejado (+, -, *, /): ")

    while True:
      try:
        n2_str = input("Digite o segundo número: ")
        n2 = float(n2_str)
        break
      except ValueError:
        print("Entrada inválida. Por favor, digite um número válido.")

    resultado = operacao(n1, n2, operador_aritmetico)

    if isinstance(resultado, str):
      print(resultado)
    else:
      print(f'O resultado de {n1} {operador_aritmetico} {n2} = {resultado:.2f}')
```

## Jogo da forca


```python
import random
import os

def clear_screen():
    # Limpa a tela do console para uma melhor experiência de jogo
    os.system('cls' if os.name == 'nt' else 'clear')

def escolher_palavra():
    palavras = [
        "python", "programacao", "desenvolvimento", "computador", "algoritmo",
        "inteligencia", "artificial", "maquinadeaprender", "dados", "colab",
        "openai", "neural", "web", "framework", "biblioteca", "funcao"
    ]
    return random.choice(palavras).upper()

def exibir_forca(tentativas_erradas):
    estagios_forca = [
        """
   ------
   |    |
   |    O
   |   /|\
   |   / \
   |
----------
        """,
        """
   ------
   |    |
   |    O
   |   /|\
   |   /
   |
----------
        """,
        """
   ------
   |    |
   |    O
   |   /|\
   |
   |
----------
        """,
        """
   ------
   |    |
   |    O
   |   /|
   |
   |
----------
        """,
        """
   ------
   |    |
   |    O
   |    |
   |
   |
----------
        """,
        """
   ------
   |    |
   |    O
   |
   |
   |
----------
        """,
        """
   ------
   |    |
   |
   |
   |
   |
----------
        """
    ]
    return estagios_forca[6 - tentativas_erradas]

def jogar_forca():
    clear_screen()
    palavra_secreta = escolher_palavra()
    letras_certas = []
    letras_erradas = []
    tentativas_erradas = 6

    print("Bem-vindo ao Jogo da Forca!")

    while True:
        print(exibir_forca(tentativas_erradas))

        palavra_exibida = ""
        for letra in palavra_secreta:
            if letra in letras_certas:
                palavra_exibida += letra + " "
            else:
                palavra_exibida += "_ "
        print(f"Palavra: {palavra_exibida}")

        if letras_erradas:
            print(f"Letras erradas: {', '.join(letras_erradas)}")
        print(f"Tentativas restantes: {tentativas_erradas}")

        if "_" not in palavra_exibida:
            print("\nParabéns! Você adivinhou a palavra!")
            break

        if tentativas_erradas == 0:
            print(f"\nQue pena! Você perdeu! A palavra era: {palavra_secreta}")
            break

        tentativa = input("Digite uma letra: ").upper()

        if not tentativa.isalpha() or len(tentativa) != 1:
            print("Por favor, digite apenas uma letra.")
            continue

        if tentativa in letras_certas or tentativa in letras_erradas:
            print(f"Você já tentou a letra '{tentativa}'. Tente outra.")
            continue

        if tentativa in palavra_secreta:
            letras_certas.append(tentativa)
            print(f"Acertou! A letra '{tentativa}' está na palavra.")
        else:
            letras_erradas.append(tentativa)
            tentativas_erradas -= 1
            print(f"Errou! A letra '{tentativa}' não está na palavra.")

        input("Pressione Enter para continuar...") # Pausa para o jogador ler a mensagem
        clear_screen()

if __name__ == "__main__":
    jogar_forca()
```

```python
import requests
import json

def consultar_cep(cep: str) -> dict | None:
  """
  Consulta informações de endereço a partir de um CEP utilizando a API ViaCEP.

  Args:
    cep: Uma string contendo o CEP a ser consultado (apenas números).

  Returns:
    Um dicionário com os dados do endereço (CEP, logradouro, complemento,
    bairro, localidade, uf) se a consulta for bem-sucedida, ou None em caso de erro.
  """
  # Remove caracteres não numéricos do CEP
  cep = ''.join(filter(str.isdigit, cep))

  if len(cep) != 8:
    print(f"Erro: O CEP '{cep}' deve conter 8 dígitos numéricos.")
    return None

  url = f"https://viacep.com.br/ws/{cep}/json/"

  try:
    response = requests.get(url, timeout=5) # Adiciona um timeout para a requisição
    response.raise_for_status() # Lança um erro HTTP para status de erro (4xx ou 5xx)

    data = response.json()

    if 'erro' in data and data['erro']:
      print(f"Erro: CEP '{cep}' não encontrado.")
      return None

    return data

  except requests.exceptions.HTTPError as http_err:
    print(f"Erro HTTP ao consultar o CEP '{cep}': {http_err}")
    return None
  except requests.exceptions.ConnectionError as conn_err:
    print(f"Erro de conexão ao consultar o CEP '{cep}': Verifique sua internet. {conn_err}")
    return None
  except requests.exceptions.Timeout as timeout_err:
    print(f"Erro de timeout ao consultar o CEP '{cep}': O servidor demorou muito para responder. {timeout_err}")
    return None
  except requests.exceptions.RequestException as req_err:
    print(f"Ocorreu um erro inesperado na requisição para o CEP '{cep}': {req_err}")
    return None
  except json.JSONDecodeError as json_err:
    print(f"Erro ao decodificar a resposta JSON da API para o CEP '{cep}': {json_err}")
    return None

# --- Exemplo de uso ---
if __name__ == "__main__":
  print("\n--- Consulta de CEP ---")
  while True:
    cep_input = input("Digite o CEP (apenas números ou 'sair' para encerrar): ")
    if cep_input.lower() == 'sair':
      break

    endereco = consultar_cep(cep_input)

    if endereco:
      print("\n--- Detalhes do Endereço ---")
      print(f"CEP: {endereco.get('cep')}")
      print(f"Logradouro: {endereco.get('logradouro')}")
      print(f"Complemento: {endereco.get('complemento')}")
      print(f"Bairro: {endereco.get('bairro')}")
      print(f"Cidade: {endereco.get('localidade')}")
      print(f"Estado: {endereco.get('uf')}")
      print("---------------------------")
    else:
      print(f"Não foi possível obter informações para o CEP '{cep_input}'.")

  print("\nConsulta de CEP encerrada.")
```
