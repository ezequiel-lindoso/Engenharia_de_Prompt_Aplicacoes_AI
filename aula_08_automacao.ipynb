```python
def verificar_temperatura(temp):
    if temp > 30:
        print(f'Alerta: Superaquecimento!({temp}°C)')
    else:
        print(f'Temperatura normal. ({temp}°C)')

temperatura_atual = float(input('Informe a temperatura do sistema: '))
verificar_temperatura(temperatura_atual)
```

#Missão 02
Desenvolver um script que consulte uma API pública (viaCEP) e retorne as informações de endereço formatada de maneira leegível.

```python
import requests

def consultar_cep(cep):
    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para respostas HTTP ruins (4xx ou 5xx)
        data = response.json()

        if 'erro' in data:
            print(f"Erro: CEP {cep} não encontrado.")
            return None
        else:
            print("--- Endereço Encontrado ---")
            print(f"CEP: {data.get('cep')}")
            print(f"Logradouro: {data.get('logradouro')}")
            print(f"Complemento: {data.get('complemento')}")
            print(f"Bairro: {data.get('bairro')}")
            print(f"Cidade: {data.get('localidade')}")
            print(f"Estado: {data.get('uf')}")
            print(f"IBGE: {data.get('ibge')}")
            return data
    except requests.exceptions.RequestException as e:
        print(f"Erro na requisição: {e}")
        return None

# Exemplo de uso:
cep_para_consultar = input("Digite o CEP para consulta: ")
consultar_cep(cep_para_consultar)
```

### Programa de Alerta Autônomo
Este programa simula o monitoramento de uma condição (temperatura do sistema) e dispara alertas quando um limite crítico é excedido. O monitoramento é feito em um loop com um atraso para simular verificações periódicas.

```python
import time
import random

def disparar_alerta(mensagem, tipo_alerta='console'):
    """Dispara um alerta, seja no console ou simulando um e-mail."""
    if tipo_alerta == 'console':
        print(f"\n!!! ALERTA CRÍTICO: {mensagem} !!!")
    elif tipo_alerta == 'email':
        print(f"\nSimulando envio de e-mail com a mensagem: {mensagem}")
        print("Para um sistema real, você integraria aqui um serviço de envio de e-mails (e.g., smtplib).")
    else:
        print(f"Tipo de alerta desconhecido: {tipo_alerta} - Mensagem: {mensagem}")

def monitorar_condicao(limite_critico, intervalo_verificacao=5):
    """Monitora uma condição e dispara alertas se o limite for atingido."""
    print(f"Iniciando monitoramento. Limite crítico: {limite_critico}°C. Verificação a cada {intervalo_verificacao} segundos.")
    print("Pressione Ctrl+C para parar o monitoramento.")

    try:
        while True:
            # Simula a leitura de uma métrica (ex: temperatura)
            # Em um cenário real, você leria isso de um sensor ou API
            temperatura_atual = random.uniform(20.0, 40.0) # Gera uma temperatura aleatória entre 20 e 40

            print(f"[{time.strftime('%Y-%m-%d %H:%M:%S')}] Temperatura atual: {temperatura_atual:.1f}°C", end='\r')

            if temperatura_atual > limite_critico:
                mensagem_alerta = f"Temperatura ({temperatura_atual:.1f}°C) excedeu o limite crítico de {limite_critico}°C!"
                disparar_alerta(mensagem_alerta, tipo_alerta='console')
                # Você pode adicionar mais tipos de alerta aqui
                # disparar_alerta(mensagem_alerta, tipo_alerta='email')

            time.sleep(intervalo_verificacao)
    except KeyboardInterrupt:
        print("\nMonitoramento interrompido pelo usuário.")

# --- Exemplo de Uso ---
# Definir o limite crítico e o intervalo de verificação
LIMITE_CRITICO = 30.0 # Temperatura em °C
INTERVALO_VERIFICACAO = 5 # Segundos

monitorar_condicao(LIMITE_CRITICO, INTERVALO_VERIFICACAO)
```
